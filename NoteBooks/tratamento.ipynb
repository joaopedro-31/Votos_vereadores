```python
vereadores = r'C:\Users\jpkab\OneDrive\Desktop\Politics_AIAgent\Fortaleza\Dados_Unificados\todos_candidatos_tratados.csv'
df = pd.read_csv(vereadores)
df_filter = df[['CANDIDATO','PARTIDO','BAIRRO','REGIONAL','VOTOS']]
df_filter = df_filter.groupby(['CANDIDATO','PARTIDO','BAIRRO','REGIONAL']).sum('VOTOS').sort_values('VOTOS', ascending=0).reset_index()
display(df_filter)
```

```python
df_bairro = df_filter.groupby(["BAIRRO", "CANDIDATO", "PARTIDO"], as_index=False)["VOTOS"].sum()
df_bairro = df_bairro.sort_values(["BAIRRO", "VOTOS"], ascending=[True, False])
df_top10_bairro = df_bairro.groupby("BAIRRO").head(10).reset_index(drop=True)

df_regional = df_filter.groupby(["REGIONAL", "CANDIDATO", "PARTIDO"], as_index=False)["VOTOS"].sum()
df_regional = df_regional.sort_values(["REGIONAL", "VOTOS"], ascending=[True, False])
df_top10_regional = df_regional.groupby("REGIONAL").head(10).reset_index(drop=True)

display(df_top10_bairro)
display(df_top10_regional)

df_top10_bairro.to_excel('top10_bairro.xlsx', index=False)
df_top10_regional.to_excel('top10_regional.xlsx', index=False)
```
