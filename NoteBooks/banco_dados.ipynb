```python
def bairro_regional(bairro: pd.Series) -> pd.Series:
    bairro_normalizado = bairro.str.strip().str.upper()
    regional = bairro_normalizado.map(dict_regionais_map)
    return regional.astype('Int64')
    
def tratar_csv(caminho_arquivo: Path) -> pd.DataFrame:
    df = pd.read_csv(caminho_arquivo, sep=';', encoding= 'utf-8')
    df.columns = [c.strip().upper() for c in df.columns]
    df = df.dropna(subset=['BAIRRO'])
    df['NÚMERO'] = df['NÚMERO'].astype('Int64')
    df['TURNO'] = df['TURNO'].astype('Int64')
    df['REGIONAL'] = bairro_regional(df['BAIRRO'])
    df['PARTIDO'] = df['NÚMERO'].apply(lambda x: dict_partidos.get(str(x)[:2], ["Desconhecido"])[0])
    colunas = ['ANO','TURNO','MUNICÍPIO','CANDIDATO','PARTIDO','VOTOS','LOCAL DE VOTAÇÃO','BAIRRO','REGIONAL']
    df = df[colunas]
    return df
```

```python
pasta_csv = Path('csvs')
arquivos_csv = list(pasta_csv.glob('*.csv'))
todos_dfs = []
csv_erro = []

for arquivo in arquivos_csv:
    try:
        df_tratado = tratar_csv(arquivo)
        todos_dfs.append(df_tratado)
    except Exception as e:
        print(f"Erro no arquivo {arquivo.name}: {e}")
        csv_erro.append(arquivo.name)

# concatenar tudo em um DataFrame único
df_final = pd.concat(todos_dfs, ignore_index=True)
print(df_final.head())
print(csv_erro)
print(df_final.shape)
df_final.to_csv("todos_candidatos_tratados.csv", index=False)
```
